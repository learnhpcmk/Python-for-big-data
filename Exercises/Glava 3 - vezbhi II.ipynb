# Пристап до податоци преку API до Скопје Пулс

Скопје Пулс е проект за мерење на загаденоста на воздухот развиен од ентизијасти. Од мрежа на сензори поставени во Скопје, но и во други градови се мери и споделува состојбата на квалитетот на воздухот. Дел од збирните и обработените податоци се достапни на веб страната на проектот, а останатите податоци од сензорите се јавно достапни преку API. Во документацијата за API достапна на https://skopje.pulse.eco/restapi е подробно опишано како се пристапува до податоците. За да пристапите потребно е да се регистрирате и со вашите корисничко име и лозинка пристапувате до она што ве интересира. При пристапот треба да се употребат вашето корисничко име и лозинка. За целта на овој курс е направено едно корисничко име, кое е дадено во кодот кој што следи.

Податоците што се добиваат се во облик JSON и треба да се проучат добро за да може да се употребат правилно. Потреба е обработка на добиените податоци за да може да се организираат според вашите потреби



```python
import requests
import pandas as pd

user = 'EuroCC'
passKurs = 'Eur0ccPA$$'
url = 'https://skopje.pulse.eco/rest/sensor'

BaranjeZaPodatoci = requests.get(url, auth=requests.auth.HTTPBasicAuth(user, passKurs))
jsonPodatoci = BaranjeZaPodatoci.json() #  Добиените податоци се всушност низа од податоци за секој сензор

for i in range(len(jsonPodatoci)):    
    train = pd.DataFrame.from_dict(jsonPodatoci[i], orient='index')
    train.reset_index(level=0, inplace=True)
    print(train)
```

Сега ќе се одбидеме да го сместиме податоците од сите сензори во една Пандас табела, за поедноставен преглед, но и за понатамошна анализа

```python
import requests
import pandas as pd

user = 'EuroCC'
passKurs = 'Eur0ccPA$$'
url = 'https://skopje.pulse.eco/rest/sensor'

BaranjeZaPodatoci = requests.get(url, auth=requests.auth.HTTPBasicAuth(user, passKurs))
jsonPodatoci = BaranjeZaPodatoci.json() #  Добиените податоци се всушност низа од податоци за секој сензор

KolkuSenzori = len(jsonPodatoci)
# print(KolkuSenzori)
EdenSenzor = pd.DataFrame.from_dict(jsonPodatoci[0], orient='index')
print(EdenSenzor[0])  # печатиме за да видиме како изгледаат податоците за првиот сензор (ред. бр. 0)
Koloni = EdenSenzor.index.values   #  насловите на колоните се земаат од првиот сензор 

SiteSenzori = pd.DataFrame(columns = Koloni)  # Празна Пандас табела со зададени наслови на колони
SiteSenzori.loc[0] = EdenSenzor.iloc[:, 0].values # Вредностите (.values) на првата колона

for i in range(len(jsonPodatoci)):    
    EdenSenzor = pd.DataFrame.from_dict(jsonPodatoci[i], orient='index') # Ги земаме податоците еден по еден
    SiteSenzori.loc[i] = EdenSenzor.iloc[:, 0].values

SiteSenzori.head()  # Проверка дали е добиена добра табела

#print(SiteSenzori)

SamoAktivni = SiteSenzori[SiteSenzori['status'] == 'ACTIVE'] # Ги земаме само активните
SamoAktivni.head(10)
```

# Дополнителна вежба за бази на податоци

Во овој дел покрај извлекувањето на една табела од базата Чанук, ќе ги назначиме и насловите на колоните

```python
import sqlite3
import pandas as pd
conn = sqlite3.connect('chinook.db')
conn.row_factory = sqlite3.Row  # Овој ред помага да се извлечат насловите на колоните
cursor = conn.cursor()
cursor.execute("SELECT * FROM employees")
IzvlechenRed = cursor.fetchone() #  Извлекува еден ред
NasloviNaKoloni = IzvlechenRed.keys() # Ги бараме клучевите за прочитаниот ред, што се всушност насловите на колоните
IzvlechenaTabela = cursor.fetchall()
PandasTabela = pd.DataFrame(IzvlechenaTabela, columns = NasloviNaKoloni)
PandasTabela.head(10) # Бараме 10 реда и се доби целата табела оти се состои од само 7 записи
```
